```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_excel('store/data_preprocessed_ver4_29Oct.xlsx')
```

```python
df = df[df['diem_hp'] <= 10]
```

```python
df = df.drop(['Unnamed: 0', 'malop', 'mamh_tt'], axis=1)
```

```python
df.columns
```

```python
label_encoders = {}
for column in ['mssv', 'mamh', 'DVQLCM']:
    le = LabelEncoder()
    df[column + '_Encoded'] = le.fit_transform(df[column])
    label_encoders[column] = le
```

```python
df.columns
```

```python
# Features for clustering
```

```python
# Features for clustering: Course_Encoded, Major_Encoded, Grade_Num, Class_Encoded
X = df[['mamh_Encoded', 'DVQLCM_Encoded', 'hocky', 'namhoc','diem_hp']] 
```

```python
# Standardizing data for clustering
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# Elbow

```python
inertia = []

# Test with group sizes from 1 to 10
for k in range(1, 30):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(range(1, 30), inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.xticks(range(1, 30))
plt.grid(True)
plt.show()
```

# Choose k = 7

# 1.Kmeans

```python
# KMeans clustering
kmeans = KMeans(n_clusters=7, random_state=42)
df['kmeans_labels'] = kmeans.fit_predict(X_scaled)
```

```python
# Plot clustering results
plt.figure(figsize=(8, 6))
sns.scatterplot(x='mamh_Encoded', y='diem_hp', hue='kmeans_labels', data=df, palette='Set1', s=100, edgecolor='k')
plt.title('KMeans Clustering of Students Based on Courses and Grades')
plt.xlabel('Mã môn học (Encoded)')
plt.ylabel('Điểm học phần')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

# 2. Gaussian Mixture Model (GMM)

```python
from sklearn.mixture import GaussianMixture
```

```python
gmm = GaussianMixture(n_components=7, random_state=42)
df['gmm_labels'] = gmm.fit_predict(X_scaled)
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='mamh_Encoded', y='diem_hp', hue=df['gmm_labels'], data=df, palette='Set2', s=100)
plt.title('GMM Clustering of Students')
plt.xlabel('Mã môn học')
plt.ylabel('Điểm học phần')
plt.grid(True)
plt.show()
```
