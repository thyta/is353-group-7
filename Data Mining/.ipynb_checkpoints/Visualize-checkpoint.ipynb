# Thư viện

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np
```

```python
fact =  pd.read_excel("..\Data Format\store\\fact.xlsx")
```

```python
fact = fact.drop(columns=["Unnamed: 0"])
```

# Xử lý lại fact dữ liệu

```python
#  ép kiểu namsinh sang int để tránh 1999.0
fact['namsinh'] = fact['namsinh'].astype(int)
```

```python
# tính toán khoá chuẩn của sinh viên ví dụ sinh năm 2003 thì khoá học chuẩn là k16 bằng cách lấy năm sinh trừ đi 1987
fact['khoahoc_chuan'] = fact['namsinh'] - 1987

fact['chenhlech'] = fact['khoahoc'] - fact['khoahoc_chuan']
```

```python
fact[['khoahoc_chuan', 'khoahoc', 'chenhlech']]
```

# Viết hàm

```python
def pie_chart(column_name, df):
    # Đếm số lượng theo column_name
    counts = fact[column_name].value_counts()

    # Vẽ biểu đồ tròn
    plt.figure(figsize=(5, 6))
    plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("deep", len(counts)))
    plt.title('Phân bố loại tốt nghiệp', fontsize=16)
    plt.axis('equal')  # Đảm bảo biểu đồ tròn
    
    # Hiển thị biểu đồ
    plt.show()
```

```python
def simple_bar_chart(column_name, display_name):
    counts = fact[column_name].value_counts()

    plt.figure(figsize=(8, 6))
    sns.barplot(x=counts.index, y=counts.values, hue=counts.index, palette="deep")
    plt.title('Phân bố ' + display_name, fontsize=16)
    plt.xlabel('số lượng', fontsize=12)
    plt.ylabel(display_name, fontsize=12)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

```python
def bar_chart_itegrate_XLHV(column_name, display_name):
    # Đếm số lượng xếp loại theo khóa học
    counts = fact.groupby([column_name, 'xeploai']).size().reset_index(name='count')

    # Vẽ biểu đồ
    plt.figure(figsize=(10, 6))
    sns.barplot(x=column_name, y='count', hue='xeploai', data=counts, palette='deep')
    plt.title('Phân bố Xếp loại theo ' + display_name, fontsize=16)
    plt.xlabel(display_name, fontsize=12)
    plt.ylabel('Số lượng', fontsize=12)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    
    # Hiển thị biểu đồ
    plt.show()

    return counts
```

# 0. Xeploai

```python
pie_chart('xeploai')
```

# 1. khoá

```python
simple_bar_chart('khoahoc', 'khoá học')
```

```python
khoahoc_xeploai_counts = bar_chart_itegrate_XLHV('khoahoc', 'khoá học')
```

```python
pie_chart('gioitinh')
```

```python
khoahoc_xeploai_counts = bar_chart_itegrate_XLHV('gioitinh', 'giới tính')
```
