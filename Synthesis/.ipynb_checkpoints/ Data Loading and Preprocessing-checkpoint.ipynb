# filter môn học tương tự

```python
mamh_tt = df[['mamh', 'mamh_tt']].dropna()

# Drop duplicate rows
mamh_tt = mamh_tt.drop_duplicates()
```

```python
mamh_tt
```

```python
dict_mamh_tt = dict(zip(mamh_tt['mamh'], mamh_tt['mamh_tt']))

df['mamh'] = df.apply(lambda row: dict_mamh_tt.get(row['mamh'], row['mamh']), axis=1)
```

```python
# Calculate value counts for 'mamh' column
mamh_counts = df[['mamh']].value_counts()

# Calculate the mean of these counts
mean_count = mamh_counts.mean()

# Filter mamh values with counts greater than the mean
mamh_above_mean = mamh_counts[mamh_counts > mean_count].reset_index()

# Rename columns for clarity
mamh_above_mean.columns = ['mamh', 'count']

# Display the resulting DataFrame
mamh_above_mean
```

```python
# Get the list of mamh values with counts above the mean
mamh_above_mean_list = mamh_above_mean['mamh'].tolist()

# Filter the original DataFrame to keep only rows where mamh is in the mamh_above_mean list
df_filtered = df[df['mamh'].isin(mamh_above_mean_list)]
```

```python
monhoc = monhoc[['Mã MH','Đơn vị quản lý chuyên môn', 'Số TC']].dropna()

dict_monhoc = dict(zip(monhoc['Mã MH'], monhoc[['Đơn vị quản lý chuyên môn', 'Số TC']].values.tolist()))

# Map the dictionary to df
df[['Đơn vị quản lý chuyên môn', 'Số TC']] = df['mamh'].map(dict_monhoc).apply(pd.Series)
```
