# Viết hàm

```python
def pie_chart(column_name, fact=fact):
    # Đếm số lượng theo column_name
    counts = fact[column_name].value_counts()

    # Vẽ biểu đồ tròn
    plt.figure(figsize=(5, 6))
    plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("deep", len(counts)))
    plt.title('Phân bố loại tốt nghiệp', fontsize=16)
    plt.axis('equal')  # Đảm bảo biểu đồ tròn
    
    # Hiển thị biểu đồ
    plt.show()
```

```python
def simple_bar_chart(column_name, display_name, fact=fact):
    counts = fact[column_name].value_counts()

    plt.figure(figsize=(8, 6))
    sns.barplot(x=counts.index, y=counts.values, hue=counts.index, palette="deep")
    plt.title('Phân bố ' + display_name, fontsize=16)
    plt.xlabel('số lượng', fontsize=12)
    plt.ylabel(display_name, fontsize=12)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

```python
def bar_chart_itegrate_XLHV(column_name, display_name, fact=fact):
    # Đếm số lượng xếp loại theo khóa học
    counts = fact.groupby([column_name, 'xeploai']).size().reset_index(name='count')

    # Vẽ biểu đồ
    plt.figure(figsize=(10, 6))
    sns.barplot(x=column_name, y='count', hue='xeploai', data=counts, palette='deep')
    plt.title('Phân bố Xếp loại theo ' + display_name, fontsize=16)
    plt.xlabel(display_name, fontsize=12)
    plt.ylabel('Số lượng', fontsize=12)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    
    # Hiển thị biểu đồ
    plt.show()

    return counts
```
